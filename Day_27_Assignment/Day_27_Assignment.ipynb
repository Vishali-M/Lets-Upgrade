# KNN

```python
df.columns
```

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df["Sex"])
df['Embarked'] = le.fit_transform(df["Embarked"])
```

```python
X = df.drop(["PassengerId", "Pclass"], axis=1)
y = df["Pclass"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
acc_scores = dict()
def KNN(k):
    print(f"\t\t K= {k}")
    knn = neighbors.KNeighborsClassifier(n_neighbors = k)
    print(f"Accuracy score based on k = {k} using score() \n{round(knn.fit(X_train,y_train).score(X_test,y_test)*100,4)}\n")

    pred = knn.predict(X_test)
    
    
    acc_score = round(accuracy_score(y_test, pred, normalize=True)*100,4)
    print(f"Accuracy score using accuracy_score() \n{acc_score}\n")
    
    conf = confusion_matrix(y_test, pred)
    lr  = conf[0, 0] + conf[1, 1]
    rl = conf[0, 1] + conf[1, 0]
#     conf_matrix_acc=round((lr/(lr+rl))*100,4)
#     print(f"Accuracy score using accuracy_score() \n{round(conf_matrix_acc*100,4)}")
    print(f"Confusion Matrix: \n{conf}")
    
    acc_scores[k] = acc_score

    
    print("*************************************************************************************")
```

```python
for x in range(1, 623):
    KNN(x)
```

```python
scores = acc_scores.values()
ks = acc_scores.keys()

max_score = max(scores)
max_k = list(ks)[list(scores).index(max_score)]

max_k, max_score
```

# SVM

```python
clf = svm.SVC(gamma = 0.01, C = 100)
def SVM(x, y):
    X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state = 0)
    y_pred = clf.fit(X_train, y_train).predict(X_test)
    acc = round(accuracy_score(y_test, y_pred, normalize = True) * 100, 2)
    confm = confusion_matrix(y_test, y_pred)
    return [acc, confm]
```

```python
df = df.drop(['PassengerId'], axis = 1)
```

```python
models = []
for enum, i in enumerate(df.columns):
    if i == 'Age' or i == 'Fare':
        continue
    y = df.iloc[:, enum:enum+1]
    x = df.drop(columns = y.columns)
    models.append([SVM(x, y), y.columns[0]])
```

```python
for _ in models:
    print(f'{_[1]} vs {str(df.drop([_[1]], axis=1).columns.to_list()).strip("[]")}\n')
    print(f'Accuracy: {_[0][0]}')
    print(f'Confusion Matrix: \n {_[0][1]}\n')
    print(f"Out of total 267 records only {np.trace(_[0][1])} are successfully classified, giving a model accuracy of {_[0][0]}%.")
    print('\n-----------------------\n')
```
